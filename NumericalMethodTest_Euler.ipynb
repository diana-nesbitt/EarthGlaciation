```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
tmax = 2
dt = 0.2
N = 10
```

```python
# ODE function
def myfunc(t, Y, ep):
    y, yp = Y
    ypp = -1/(1 + ep*y)**2
    return np.array([yp, ypp])
# Because solve_ivp wants a function with only two inputs, create an auxiliary function to pass any arguments
ep = 0.2
f = lambda t, Y: myfunc(t, Y, ep)
```

```python
# initial condition
Y0 = [0, 1]

Y = Y0
t = 0
for i in range(0, N):
    t = t + dt
    #print(Y+f(t,Y)*dt)
    Y = Y + f(t, Y)*dt
    plt.plot(t, Y[0], 'bo')
    
tt = np.linspace(0, N*dt, 100)
y0 = -1/2*tt**2 + tt
plt.plot(tt, y0, '--')
```
