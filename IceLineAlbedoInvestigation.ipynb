# Steady State - Effect of parameters on the ice-age state

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ys = 0.9 # this is the ice boundary determined by the model. Our ice boundary is "close to the north pole" - let's say 0.9
y = np.array([0, ys-0.0001, ys, ys+0.0001, 1])
ai = 0.62; aw = 0.32; 
a = np.piecewise(y, [y < ys, y == ys, y > ys], [aw, 0.5*(aw+ai), ai])
plt.plot(y, a, '-o') 
plt.xlabel('y'); plt.ylabel('a(y)');
```
