```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
ep = np.linspace(1, 0.1, 20)
root1 = (-1 + np.sqrt(1 - 4*ep*-1))/(2*ep)
root2 = (-1 - np.sqrt(1 - 4*ep*-1))/(2*ep)
asym1 = 1 - ep
asym2 = -1/ep - 1
plt.plot(ep, root1, 'o')
plt.plot(ep, root2, 'o')
plt.plot(ep, asym1)
plt.plot(ep, asym2)
plt.xlabel('$\epsilon$')
plt.ylabel('x')
```
